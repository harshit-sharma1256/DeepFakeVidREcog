```python
face_detector = MTCNN()
```

```python
def determining_angle_of_eyes(left_eye_kp,right_eye_kp):

  del_y = right_eye_kp[1] - left_eye_kp[1]
  del_x = right_eye_kp[0] - left_eye_kp[0]

  return np.degrees(np.arctan2(del_y,del_x))
```

```python
def determine_scaling_factor(og_img_left_eye_kp,og_img_right_eye_kp,rescaled_img_left_eye_kp,
                             rescaled_img_right_eye_kp):
  
  del_y_og = og_img_right_eye_kp[1] - og_img_left_eye_kp[1]
  del_x_og = og_img_right_eye_kp[0] - og_img_left_eye_kp[0]

  eye_dist_og = np.sqrt((del_y_og**2) + (del_x_og**2))

  del_y_rescaled = rescaled_img_right_eye_kp[1] - rescaled_img_left_eye_kp[1]
  del_x_rescaled = rescaled_img_right_eye_kp[0] - rescaled_img_left_eye_kp[0]

  eye_dist_rescaled = np.sqrt((del_y_rescaled)**2 + (del_x_rescaled)**2)

  scaling_factor = eye_dist_rescaled/eye_dist_og

  return scaling_factor
```

```python
def crop_faces_and_realign(cropped_face_img_size,eye_loc,og_img,left_eye_kp,right_eye_kp):

  rescaled_img_left_eye_kp = (eye_loc[0]*cropped_face_img_size, eye_loc[1]*cropped_face_img_size)
  rescaled_img_right_eye_kp = ((1-eye_loc[0])*cropped_face_img_size, eye_loc[1]*cropped_face_img_size)
  #print(rescaled_img_left_eye_kp,rescaled_img_right_eye_kp)

  eyes_center = ((left_eye_kp[0] + right_eye_kp[0])//2, (left_eye_kp[1] + right_eye_kp[1])//2)
  #print(eyes_center)

  scaling_factor = determine_scaling_factor(left_eye_kp,right_eye_kp,rescaled_img_left_eye_kp,
                                            rescaled_img_right_eye_kp)
  #print(scaling_factor)
  
  eyes_angle = determining_angle_of_eyes(left_eye_kp,right_eye_kp)
  #print(eyes_angle)

  rotation_matrix = cv2.getRotationMatrix2D(center=eyes_center,angle=eyes_angle,scale=scaling_factor)
  #print(rotation_matrix)

  #print(rescaled_img_width,rescaled_img_left_eye_kp[1])

  translated_x = cropped_face_img_size * 0.5
  translated_y = rescaled_img_left_eye_kp[1]
  #print(rescaled_img_width,rescaled_img_left_kp[1])

  rotation_matrix[0,2] = rotation_matrix[0,2] + (translated_x - eyes_center[0])
  rotation_matrix[1,2] = rotation_matrix[1,2] + (translated_y - eyes_center[1])

  aligned_and_rescaled_face_img = cv2.warpAffine(src=og_img,M=rotation_matrix,
                                              dsize=(cropped_face_img_size,cropped_face_img_size))
  
  #print(rotated_and_scaled_image)

  return aligned_and_rescaled_face_img
```

```python
def detect_single_image_face(og_img,eye_loc,cropped_face_img_size):

  rgb_img = cv2.cvtColor(og_img,cv2.COLOR_BGR2RGB)
  face_attr = face_detector.detect_faces(rgb_img)
  #print(result)

  left_eye_kp = face_attr[0]["keypoints"]["left_eye"]
  right_eye_kp = face_attr[0]["keypoints"]["right_eye"]

  #print(rescaled_img_width,eye_loc_percentage,left_eye_kp,right_eye_kp)

  aligned_and_rescaled_face_img = crop_faces_and_realign(cropped_face_img_size,eye_loc,og_img,left_eye_kp,right_eye_kp)

  return aligned_and_rescaled_face_img
```

```python
dst_face_vid_path = "/content/preprocessed_deepfake_vids"
```

```python
def preprocess_each_vid_frame(vid_path,vid_counter,dst_face_vid_path):
  
  print(f"Processing video: {vid_path}")

  video = cv2.VideoCapture(str(vid_path))

  cropped_faces_np_array_list = list()

  while True:

    ret,frame = video.read()

    if not ret:
      break

    aligned_cropped_rescaled_face_np_array = detect_single_image_face(og_img=frame,
                                                                      eye_loc=(0.375,0.375),
                                                                      cropped_face_img_size=224)

    cropped_faces_np_array_list.append(aligned_cropped_rescaled_face_np_array)

  video.release()

  np.savez(dst_face_vid_path+"/face_vid"+str(vid_counter)+".npz",np.array(cropped_faces_np_array_list))

  print("Saved .npz array as "+dst_face_vid_path+"/face_vid"+str(vid_counter)+".npz")
```

```python
for single_dir,single_face_dir in zip(og_deepfake_vids_path_list,
                                                og_deepfake_face_vids_path_list):
  
  start_vid_counter = 0

  for dir_path in pathlib.Path(single_dir).iterdir():
    if not dir_path.is_dir():
        continue

    for single_dir_path in dir_path.glob("*.avi"):

      end_vid_counter = start_vid_counter + 1

      with ThreadPoolExecutor(max_workers=4) as pool:
        pool.submit(preprocess_each_vid_frame,single_dir_path, start_vid_counter,
               single_face_dir)

      start_vid_counter = end_vid_counter
```
