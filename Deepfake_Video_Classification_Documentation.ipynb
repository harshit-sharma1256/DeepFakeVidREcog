###Deepfakes are videos in which the face of a person is automatically swapped with the face of another person, typically using a pretrained Generative Adversarial Network (GAN). Deepfake videos are looking more and more realistic, and it is becoming increasingly difficult to differentiate the real video from a deepfake. Therefore, many people are designing and developing systems for the detection of deepfakes.

###To know and see how realistic deepfake videos have become in recent years and how they can be a big threat in variety of domains, see the [video](https://www.youtube.com/watch?v=4vB4VL6pfHk) of a politician where several deepfake versions are shown.

###Therefore, it can be easily said that the task of differentiating a Real Video (Class Label 0) from a Deepfake one (Class Label 1) is a Binary Classification task. In order to achieve this task with as much good Performance Metrics (Precision, Recall, F1-Score, Accuracy) as possible, we will be training a Deep Learning based system which will take a Video as input and give the output as $P(Class=Fake|X_i)$ where $X_i$ is a random $i^{th}$ video which is input to the Deep Learning based System.

###Each video can be seen as a collection of lot of frames (Images). If a video is black and white then each frame will be a grayscale image which will be a ```np.array``` having shape as ```(rows,columns)``` whereas if it is a coloured video, then each frame will be coloured RGB image which will a ```np.array``` having shape as ```(rows,columns,3)```, as shown below: 

![](https://drive.google.com/uc?id=1aozfMcvi-7tgxyfvurC-Zcw3e7HeBlbx)

###Usual Frame rate of a video is 30 fps (Frames Per Second), which means that if we have a video of 1 Second duration then that video will be a collection of 30 Frames (30 Images) and therefore, just like a single image can be represented as a ```np.array```, a Grayscale (Coloured) Video can again be represented as a ```np.array``` of shape ```(30, rows, columns)``` or ```(30, rows, columns, 3)```. 

###Now, we have to do data preprocessing of these videos so that we can make our videos ready for training a deep learning based system. Finally, we will save each video in a compressed NumPy array format (```.npz``` format) in our linux instance drive:
